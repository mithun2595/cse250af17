```python
from matplotlib import pyplot as plt
import math

def gx(x):
    return sum([math.tanh(x+(2/(d**0.5))) for d in range(1,11)])/10

def fx(x):
    return math.tanh(x)

def aux_converge(x0, func):
    x_val = [x0]
    x_val.append(x_val[len(x_val)-1] - func(x_val[len(x_val)-1]))
    while x_val[len(x_val)-1] != x_val[len(x_val)-2]:
        x_val.append(x_val[len(x_val)-1] - func(x_val[len(x_val)-1]))
    plt.figure(figsize=(6,6))
    plt.plot(x_val)
    plt.title('Converges at '+str(x_val[len(x_val)-1]),fontsize=14)
    plt.xlabel('Iterations', fontsize=14)
    plt.ylabel('Xn', fontsize=14)
    plt.grid(True)
    plt.show()
```

```python
aux_converge(-2, fx)
```

```python
aux_converge(3,fx)
```

```python
aux_converge(10,gx)
```
